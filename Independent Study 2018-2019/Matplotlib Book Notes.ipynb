# Matplotlib Bible

```python
#Imports matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
#classic style
plt.style.use('fivethirtyeight')
#%matplotlib notebook is interactive plots
#%matplotlib inline is static plots
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
#0-10 with 100 elements
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x),'--')
#shows the plot
plt.show()
#fig.savefig('my_figure.png')
```

```python
#create a plot figure
plt.figure()
#create the first of two panels and set current axis
ax1 = plt.subplot(2,1,1)
ax1.plot(x, np.sin(x))
#create the second panel and set current axis
ax2 = plt.subplot(2,1,2)
ax2.plot(x,np.cos(x))
plt.show()
#subplot(nrows, ncols, index, **kwargs)
```

```python
#manually call figure, axis
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x),color='green',linestyle='-.')
#can also do single nonkeyword argument
#ex: ax.plot(x,np.sin(x),'--g')
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
#reversing axis is just switching order of arguments
#set x,y limits in single call!
plt.axis([-1,11,-1.5,1.5])
#auto tightens bounds- plt.axis('tight')
#equal aspect ratio- plt.axis('equal')
plt.show()
```

```python
plt.plot(x,np.sin(x))
plt.title('A Sine Curve')
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
plt.plot(x,np.sin(x),'-g',label='sin(x)')
plt.plot(x,np.cos(x),':b',label='cos(x)')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
y = np.linspace(0,10,30)
plt.plot(y,np.cos(y),'o',color='black')
plt.show()
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker, label = "marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0,1.8)
plt.show()
```

```python
rng = np.random.RandomState(0)
h = rng.randn(100)
v = rng.randn(100)
colors = rng.randn(100)
sizes = 1000* rng.rand(100)
plt.scatter(h,v,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
#plt.plot is more efficent than plt.scatter
#plt.scatter renders each point a separate size and color
```

```python
o = np.linspace(0,10,50)
dy = 0.8
p = np.sin(o) + dy * np.random.randn(50)
plt.errorbar(o,p,yerr=dy,fmt='.',color='black',ecolor='lightgray',elinewidth=1,capsize=0)
```

```python
import pandas as pd
cities = pd.read_csv('california_cities.csv')
#extract data
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
#scatter points using size, color, but no label
plt.scatter(lon,lat,label=None, c=np.log10(population),cmap='viridis',s=area,linewidth=0,alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3,7)
for area in [100,300,500]:
    plt.scatter([],[],c='k',alpha=0.3,s=area,label=str(area)+'km$^2$')
plt.legend(scatterpoints=1,frameon=False,labelspacing=1,title='City Area')
plt.title('California Cities: Area and Population')    
```
